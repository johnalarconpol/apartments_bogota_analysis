```python
# Ahora puedes contar los nulos correctamente
print(df.isnull().sum())
def extraer_numero_general(x):
    if isinstance(x, dict):
        for clave in ['$numberDouble', '$numberInt', '$numberLong']:
            if clave in x:
                valor = x[clave]
                try:
                    return float(valor) if valor.lower() != 'nan' else np.nan
                except:
                    return valor
    return x


# Aplicar la función a todas las columnas
df = df.applymap(extraer_numero_general)
```

## Imputar

- El valor '0' en *banos* significa nulo

Cambiar los valores nulos por la moda

```python
moda = df['website'].mode()[0]
df['website'] = df['website'].fillna(moda)
moda = df['last_view'].mode()[0]
df['last_view'] = df['last_view'].fillna(moda)
moda = df['datetime'].mode()[0]
df['datetime'] = df['datetime'].fillna(moda)
```

Imputar el barrio por medio de las coordenadas geograficas


```python
# Redondear coordenadas para mejorar coincidencias (ajusta la precisión si es necesario)
df['latitud_r'] = df['latitud'].round(3)
df['longitud_r'] = df['longitud'].round(3)

# Función para imputar barrio con múltiples criterios
def imputar_barrio(row, df_ref):
    if pd.notnull(row['barrio']):
        return row['barrio']

    # Intentar con localidad + estación TM
    filtro1 = (df_ref['localidad'] == row['localidad']) & \
              (df_ref['estacion_tm_cercana'] == row['estacion_tm_cercana'])

    barrios_comunes = df_ref.loc[filtro1 & df_ref['barrio'].notnull(), 'barrio'].mode()
    if not barrios_comunes.empty:
        return barrios_comunes[0]
    
    # Intentar con localidad + parque cercano
    filtro2 = (df_ref['localidad'] == row['localidad']) & \
              (df_ref['parque_cercano'] == row['parque_cercano'])

    barrios_comunes = df_ref.loc[filtro2 & df_ref['barrio'].notnull(), 'barrio'].mode()
    if not barrios_comunes.empty:
        return barrios_comunes[0]
    
    # Intentar con coordenadas redondeadas
    filtro3 = (df_ref['latitud_r'] == row['latitud_r']) & \
              (df_ref['longitud_r'] == row['longitud_r'])

    barrios_comunes = df_ref.loc[filtro3 & df_ref['barrio'].notnull(), 'barrio'].mode()
    if not barrios_comunes.empty:
        return barrios_comunes[0]

    # Si todo falla, usar la moda general
    return df_ref['barrio'].mode().iloc[0]

# Aplicar la función
df['barrio'] = df.apply(lambda row: imputar_barrio(row, df), axis=1)

# Eliminar columnas auxiliares si no las necesitas
df.drop(columns=['latitud_r', 'longitud_r'], inplace=True)

```

Imputar el sector por medio de las variables que me ayudan con la ubicacion

```python
import pandas as pd
import numpy as np

# Redondear coordenadas (opcional, para agrupar ubicaciones similares)
df['lat_r'] = df['latitud'].round(3)
df['lon_r'] = df['longitud'].round(3)

# Función para imputar sector
def imputar_sector(row, df_ref):
    if pd.notnull(row['sector']):
        return row['sector']
    
    # 1. localidad + barrio
    f1 = (df_ref['localidad'] == row['localidad']) & (df_ref['barrio'] == row['barrio'])
    moda = df_ref.loc[f1 & df_ref['sector'].notna(), 'sector'].mode()
    if not moda.empty:
        return moda[0]
    
    # 2. localidad + estación TM cercana
    f2 = (df_ref['localidad'] == row['localidad']) & (df_ref['estacion_tm_cercana'] == row['estacion_tm_cercana'])
    moda = df_ref.loc[f2 & df_ref['sector'].notna(), 'sector'].mode()
    if not moda.empty:
        return moda[0]
    
    # 3. localidad + parque cercano
    f3 = (df_ref['localidad'] == row['localidad']) & (df_ref['parque_cercano'] == row['parque_cercano'])
    moda = df_ref.loc[f3 & df_ref['sector'].notna(), 'sector'].mode()
    if not moda.empty:
        return moda[0]
    
    # 4. Coordenadas redondeadas
    f4 = (df_ref['lat_r'] == row['lat_r']) & (df_ref['lon_r'] == row['lon_r'])
    moda = df_ref.loc[f4 & df_ref['sector'].notna(), 'sector'].mode()
    if not moda.empty:
        return moda[0]
    
    # 5. Moda general
    return df_ref['sector'].mode().iloc[0]

# Aplicar la función a todas las filas
df['sector'] = df.apply(lambda row: imputar_sector(row, df), axis=1)

# Eliminar columnas auxiliares
df.drop(['lat_r', 'lon_r'], axis=1, inplace=True)
```

```python
import numpy as np

# Definir el umbral máximo razonable para un apartamento en Colombia (7.000 millones COP)
umbral_precio_maximo = 7e9  # 7,000,000,000 COP
umbral_precio_minimo = 2e7  # 20,000,000 COP
# Mostrar cuántos registros superarían ese umbral antes de limpiar
n_outliers = df[df['precio_venta'] > umbral_precio_maximo].shape[0]
print(f"Registros con precio_venta mayor a {umbral_precio_maximo:,.0f}: {n_outliers}")
# Mostrar cuántos registros superarían ese umbral antes de limpiar
n_outliers2 = df[df['precio_venta'] < umbral_precio_minimo].shape[0]
print(f"Registros con precio_venta mayor a {umbral_precio_minimo:,.0f}: {n_outliers2}")

# Reemplazar valores superiores al umbral con NaN
df.loc[df['precio_venta'] > umbral_precio_maximo, 'precio_venta'] = np.nan
df.loc[df['precio_venta'] < umbral_precio_minimo, 'precio_venta'] = np.nan
# Confirmar limpieza
print("Limpieza completada. Estadísticas actualizadas:")
print(df['precio_venta'].describe())
```
